# selecting columns

confusing syntax!
* query df with 
    * df[b] a boolean vector b of length n to select rows
    * df['foo'] a column name 'foo' to select one column
    * df[['foo','bar']] a list of column names to select several columns
* query df.iloc with
    * df.iloc[l] to select the rows with indices in the list l
    * df.iloc[:, l] to select the columns with indices in the list l

```python
# select first three columns
df.iloc[:, :3]
```

# Plot sales over time

```python
# Aggregate all sales by month and product line
sales_by_productline = df2003.groupby(['cum_month', 'PRODUCTLINE']).sum()
sales_by_productline.head()
```

```python
# fix the hierarchical index 
sales_by_productline = sales_by_productline.reset_index()
sales_by_productline.head()
```

```python
# change orderdate to a datetime using apply
def cum_month_to_datetime(cum_month):
    year = (cum_month - 1) // 12
    month = (cum_month - 1) % 12 + 1
    return pd.to_datetime(f'{year}-{month}-01')

sales_by_productline['ORDERDATE'] = sales_by_productline['cum_month'].apply(cum_month_to_datetime)
sales_by_productline.head()
```

```python
# Create a line plot of sales over time for each product category
sns.lineplot(x='ORDERDATE', y='SALES', hue='PRODUCTLINE', data=sales_by_productline)
```
